# Flight Pricing Notebook 01

# Exploration & Nettoyage

L'objectif est de comprendre la structure du dataset, assurer la qualité des données et les préparer pour l'analyse.

**Plan :**
1. Chargement général du dataset
2. Statistiques descriptives
3. Valeurs manquantes & doublons
4. Distribution des variables clés
5. Export du dataset nettoyé

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Palette de couleur pour le projet
COLORS = {
    'primary':   '#1B4F8A',
    'secondary': '#E8903A',
    'accent':    '#2ECC71',
    'danger':    '#E74C3C',
    'light':     '#ECF0F1',
    'dark':      '#2C3E50'}

PALETTE = [COLORS['primary'], COLORS['secondary'], COLORS['accent'], COLORS['danger']]

plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

print('Librairies chargées')
```

## 1. Chargement des données

```python
df = pd.read_csv('../Portfolio-Pricing-Flights/Clean_Dataset.csv')

print(f'Shape : {df.shape[0]:,} lignes × {df.shape[1]} colonnes')
df.head()
```

```python
# Types et infos générales
df.info()
```

## 2. Statistiques descriptives

```python
# Variables numériques
df.describe().round(2)
```

```python
# Variables catégorielles
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print(f'\n{col} ({df[col].nunique()} valeurs uniques):')
    print(df[col].value_counts().to_string())
```

## 3. Qualité des données

```python
# Valeurs manquantes
manquants = df.isnull().sum()
manquants_pct = (manquants / len(df) * 100).round(2)

manquants_df = pd.DataFrame({
    'Valeurs manquantes': manquants,
    '% manquants': manquants_pct
}).sort_values('% manquants', ascending=False)

print(manquants_df[manquants_df['Valeurs manquantes'] > 0])
print(f'\nColonnes sans valeurs manquantes : {(manquants == 0).sum()}/{len(df.columns)}')
```

```python
# Doublons
doublons = df.duplicated().sum()
print(f'Doublons : {doublons:,} ({doublons/len(df)*100:.2f}%)')

# Suppression doublons si besoin
if doublons > 0:
    df = df.drop_duplicates()
    print(f'Doublons supprimés. Shape : {df.shape}')
```

## 4. Distribution des variables clés

```python
# Distribution du prix qui est la variable cible
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution brute
axes[0].hist(df['price'], bins=80, color=COLORS['primary'], alpha=0.85, edgecolor='white')
axes[0].set_title('Distribution des prix (brute)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Prix (INR)')
axes[0].set_ylabel('Fréquence')
axes[0].axvline(df['price'].median(), color=COLORS['secondary'], linestyle='--', 
                linewidth=2, label=f"Médiane : {df['price'].median():,.0f}")
axes[0].legend()

# Distribution log (pour mieux voir la forme)
axes[1].hist(np.log1p(df['price']), bins=80, color=COLORS['accent'], alpha=0.85, edgecolor='white')
axes[1].set_title('Distribution des prix (log)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('log(Prix + 1)')
axes[1].set_ylabel('Fréquence')

plt.suptitle('Variable cible Prix des billets', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/01_distribution_prix.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nStats prix :")
print(f"  Moyenne   : {df['price'].mean():,.0f} INR")
print(f"  Médiane   : {df['price'].median():,.0f} INR")
print(f"  Min       : {df['price'].min():,.0f} INR")
print(f"  Max       : {df['price'].max():,.0f} INR")
print(f"  Std       : {df['price'].std():,.0f} INR")
```

```python
# Répartition par compagnie aérienne
nb_airline = df['airline'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Volume de billets
bars = axes[0].bar(nb_airline.index, nb_airline.values, 
                   color=PALETTE * 3, edgecolor='white', linewidth=0.5)
axes[0].set_title('Volume de billets par compagnie', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Compagnie')
axes[0].set_ylabel('Nombre de billets')
axes[0].tick_params(axis='x', rotation=30)
for bar, val in zip(bars, nb_airline.values):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100, 
                 f'{val:,}', ha='center', fontsize=9)

# Prix médian par compagnie
prix_airline = df.groupby('airline')['price'].median().sort_values(ascending=False)
bars2 = axes[1].bar(prix_airline.index, prix_airline.values,
                    color=PALETTE * 3, edgecolor='white', linewidth=0.5)
axes[1].set_title('Prix médian par compagnie (INR)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Compagnie')
axes[1].set_ylabel('Prix médian (INR)')
axes[1].tick_params(axis='x', rotation=30)
axes[1].axhline(df['price'].median(), color=COLORS['dark'], linestyle='--', 
                linewidth=1.5, alpha=0.6, label='Médiane globale')
axes[1].legend()

plt.suptitle('Compagnies aériennes', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/01_compagnies.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Répartition Classe économique vs Business
classe_distri = df['class'].value_counts()
classe_prix = df.groupby('class')['price'].describe().round(0)

print('📊 Répartition des classes :')
print(classe_distri.to_string())
print(f'\n💰 Prix par classe :')
print(classe_prix[['mean', '50%', 'min', 'max']].to_string())

ratio = classe_prix.loc['Business', 'mean'] / classe_prix.loc['Economy', 'mean']
print(f'\n📌 Rapport Business/Économie : x{ratio:.1f}')
```

## 5. Export du dataset nettoyé

```python
# Vérification finale avant export
print('Résumé final :')
print(f'  Lignes      : {len(df):,}')
print(f'  Colonnes    : {df.shape[1]}')
print(f'  NaN totaux  : {df.isnull().sum().sum()}')
print(f'  Doublons    : {df.duplicated().sum()}')

# Export
df.to_csv('../Portfolio-Pricing-Flights/flights_clean.csv', index=False)
print('\nDataset nettoyé exporté → data/flights_clean.csv')
```
